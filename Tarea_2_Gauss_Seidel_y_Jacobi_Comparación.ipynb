```
#Tarea 2 Comparacion Gauss Seidel y Jacobi
#Metodo Gauss Seidel

"""Observaciones: Se pudo apreciar que los puntos de iteracion en la grafica de Jacobi son los mismos con respecto
 a (x,w) y (y,z), mientras que en Gauss Seidel, los puntos varian un poco, ya que en las sustituciones se toman 
 los valores mas recientes obtenidos, y en Jacobi solo se utilizan en las sustituciones los valores anteriores obtenidos."""

import matplotlib.pyplot as plt
import math
def Gauss():
  x1=[0]
  x=0
  y1=[0]
  y=0
  z1=[0]
  z=0
  w1=[0]
  w=0
  limit=0

  def x2(y):
    return  0.25+(0.25*y)
  def y2(x,z):
    return  0.25+(0.25*x)+(0.25*z)
  def z2(y,w):
    return  0.25+(0.25*y)+(0.25*w)
  def w2(z):
    return  0.25+(0.25*z)

  while(limit <10):
    
    x = x2(y)
    x1.append(x)
    y = y2(x,z)
    y1.append(y)
    z = z2(y,w)
    z1.append(z)
    w = w2(z)
    w1.append(w)
    limit+=1
  
  for x in x1:
      print("Iteraciones X: ", x)
  for y in y1:
      print("Iteraciones y: ", y)
  for z in z1:
      print("Iteraciones z: ", z)
  for w in w1:
      print("Iteraciones w: ", w)

  plt.plot(x1, label="x")
  plt.plot(y1, label="y")
  plt.plot(z1, label="z")
  plt.plot(w1, label="w")
  plt.legend()
  plt.show()

Gauss()
```
